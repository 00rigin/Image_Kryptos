# encoding text to image


## Change to ascii Function
change text to ascii code

```python
def change_to_ascii(string, width, hor):
    change_string = []
    temp = []
    for i in range(0, len(string), width//hor):
        for j in range(i, i + width//hor):
            if j>=len(string): break
            temp.append(ord(string[j]))  
        change_string.append(temp)
        temp = []
    return change_string
```

# encording Function

```python
#incording(string, brfore_image_path, save_image_path, hor, ver)
def encording(string, hor, ver):
    before_image_path = ''
    save_image_path = ''
    img = cv.imread(before_image_path)
    #img = cv.imread(brfore_image_path)
    len_ = len(string)
    len_num = []
    a = len_
    while a>=10:
        x = a%10
        len_num.append(x)
        a = a//10
    len_num.append(a)
    len_num.reverse()

    count = 0
    height = img.shape[0]
    width = img.shape[1]
    change_string = change_to_ascii(string, width, hor)
    if len(change_string)>height//hor+hor:
        print("image is too small for encording")
        sys.exit(0)
        
     # 첫줄에는 string 길이 10 자리수
    t = 0
    for y in range(0, height, ver):
        i = 0
        if count>= len_: break
        #for x in range(1, width+1, hor):     # 값 변경 필요
        for x in range(0, width, hor):
            if y == 0:                          # 첫줄은 string의 길이 자리수별로 나눠서 들어있음
                if t>=len(len_num): break
                img.itemset(y, x, 0, len_num[t]) # blue -> when use green/red image
                #img.itemset(y, x, 1, len_num[t]) # green
                #img.itemset(y, x, 2, len_num[t]) # red
                t+=1
                
            else:
                if i != 0 and i%(width//hor)==0: break
                if count>= len_: break
                img.itemset(y, x, 0, change_string[y//ver-1][i])
                #img.itemset(y, x, 1, change_string[y//ver-1][i])
                #img.itemset(y, x, 2, change_string[y//ver-1][i])
                count+=1
                i += 1

    cv.imwrite(save_image_path, img)
    #cv.imwrite(save_image_path, img)
    cv.imshow('encorded', img)
    cv.waitKey(0)
```
